```python
import pandas as pd
import numpy as np
```

```python
stopwords=[]
```

```python
data=pd.read_csv('환경오염_불용어_2번.csv',encoding='utf-8-sig',index_col=0)
```

```python
data.head()
```

```python
t=len(data)
print(t)
```

```python
data['Unnamed: 3']==0
```

```python
for i in range(0,t):
    if data['Unnamed: 3'][i]==0:
        stopwords.append(data['명사'][i])
```

```python
len(stopwords)
```
