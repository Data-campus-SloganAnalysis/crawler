```python
tokenizer = Mecab()
pos_temp = {}
pos_list = []

for i in df['text']:
    temp_str = ''
    try:
        for j in range(len(tokenizer.pos(i))):
            pos_temp[tokenizer.pos(i)[j][1]]+=1
            temp_str+=tokenizer.pos(i)[j][1]+"/"
    except:
        for j in range(len(tokenizer.pos(i))):
            pos_temp[tokenizer.pos(i)[j][1]]=1
            temp_str+=tokenizer.pos(i)[j][1]+"/"
    pos_list.append(temp_str)
```
