```python
import re
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd


options = webdriver.ChromeOptions()
options.add_argument("headless")
```

```python
driver_path = 'C:/Users/chaeh/chromedriver.exe' 
```

```python
def clean_text(text):
    cleaned_text = re.sub('[a-zA-Z]', '', text)
    cleaned_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]',
                          '', cleaned_text)
    cleaned_text = ' '.join(cleaned_text.split())
    return cleaned_text
```

```python
def fmkorea_document(query,driver_path,page):
    driver = webdriver.Chrome(executable_path=driver_path,options=options)
    data_list=[]

    for i in range(1,page+1):
        base_url = 'https://www.fmkorea.com/?act=IS&is_keyword='
        parm = query+'&mid=home&where=document&page='+str(i)
        url = base_url + parm
        driver.get(url)
        soup = BeautifulSoup(driver.page_source,'html.parser')
        for j in range(1,11):
            regex="\[[^]]*\]"
            title=soup.select('#content > div > ul.searchResult > li:nth-of-type('+str(j)+') > dl > dt > a')
            title=title[0].text
            title=re.sub(regex,'',str(title))
            title=clean_text(title)
            
            contents=soup.select('#content > div > ul.searchResult > li:nth-of-type('+str(j)+') > dl > dd')
            contents=clean_text(contents[0].text)
            data_list.append(title+contents)
    driver.close()

    return data_list
```

```python
fmkorea_document('풀무원',driver_path,60)
```

```python
def fmkorea_comment(query,driver_path,page):
    driver = webdriver.Chrome(executable_path=driver_path,options=options)
    data_dic={}

    for i in range(1,page+1):
        base_url = 'https://www.fmkorea.com/?act=IS&is_keyword='
        parm = query+'&mid=home&where=comment&page='+str(i)
        url = base_url + parm
        driver.get(url)
        soup = BeautifulSoup(driver.page_source,'html.parser')
        for j in range(1,11):
            regex="\<[^)]*\>"
            
            contents=soup.select('#content > div > ul.searchResult.comment > li:nth-of-type('+str(j)+') > dl > dt > a')
            contents=str(contents[0])
            contents=contents.split('<br/>')
            contents[0]=re.sub(regex,'',str(contents[0]))
            regex="\[[^]]*\]"
            comment_title=re.sub(regex,'',contents[0])
            comment_title=clean_text(comment_title)
            comment_doc=clean_text(contents[1])
            if comment_title in data_dic:
                data_dic[comment_title].append(comment_doc)
            else:
                data_dic[comment_title]=[comment_doc]
    driver.close()

    return data_dic
```

```python
fmkorea_comment('풀무원',driver_path,60)
```
