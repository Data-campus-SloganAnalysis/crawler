```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
def biz_khan(query,driver_path,page):
    current_page = 1
    last_page =(int(page)-1)*10+1
    output = pd.DataFrame(columns=['언론사', '제목', '내용'])
    
    while current_page < last_page:
    
        driver = webdriver.Chrome(executable_path=driver_path)
        url_domain = 'https://search.naver.com/search.naver?where=news&query='
        url = url_domain + query +'&%2Ca%3A&start='+str(current_page)
        driver.get(url)
        soup = BeautifulSoup(driver.page_source,'html.parser')

        news_list = soup.select('#main_pack > section > div > div.group_news > ul')
        url_list = []
        news_list = news_list[0].find_all('li')


        for url in news_list:
            naver_links = url.find_all('a')
            for one_link in naver_links:
                temp_link = one_link.get('href')
                if temp_link.startswith("https://news.naver.com"):
                    url_list.append(temp_link)


        for url in url_list:
            driver.get(url)        
            soup = BeautifulSoup(driver.page_source,'html.parser')
            temp_data = {
                '언론사':soup.select('#footer address')[0].a.get_text(),
                '제목':soup.select('h3#articleTitle')[0].text,
                '내용':soup.select('#articleBodyContents')[0].text.replace('\n','')
            }
            output = output.append(temp_data,ignore_index=True)
        current_page+=10
    return output    
    driver.close()
```

```python
driver_path = './chromedriver'


# 제목, 크롬 경로, 크롤링 할 페이지 수

biz_khan('삼성 슬로건',driver_path,5)
```
